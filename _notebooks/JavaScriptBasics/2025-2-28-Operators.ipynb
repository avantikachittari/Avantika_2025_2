# **Operators in JavaScript**

## **Length of a String**
Get the number of characters in a string using.length:.
<pre><code>
let name = "Avantika";
console.log(name.length);  // Output: 8
</code></pre>

## **Substring**
Extract part of a string with .substring():
<pre><code>
let text = "Programming";
let subText = text.substring(0, 5);
console.log(subText);  // Output: Progr
</code></pre>

## **Upper and Lower Case**
Convert strings to upper or lower case:
<pre><code>
let phrase = "Hello World!";
console.log(phrase.toUpperCase());  // Output: HELLO WORLD!
console.log(phrase.toLowerCase());  // Output: hello world!
</code></pre>

## **Mathematical Expressions**
Addition (+)
Add two numbers:
<pre><code>
let a = 5;
let b = 3;
console.log(a + b);
</code></pre>

Subtraction (-)
Subtract one number from another:
<pre><code>
let a = 9;
let b = 4;
console.log(a - b);  // Output: 5
</code></pre>

Multiplication (*)
Multiply two numbers:
<pre><code>
let a = 6;
let b = 7;
console.log(a * b);  // Output: 42
</code></pre>

Division (/)
Divide one number by another:
<pre><code>
let a = 8;
let b = 2;
console.log(a / b);  // Output: 4
</code></pre>

## **Boolean Expressions**

Comparison Operators
Compare two values:
<pre><code>
let x = 10;
let y = 20;
console.log(x == y);  // Output: false
console.log(x < y);   // Output: true
</code></pre>

Logical Operators
Combine conditions:
<pre><code>
let a = 5;
let b = 10;
console.log(a < b && b > a);  // Output: true
</code></pre>
